```python
import numpy as np 
from matplotlib import pyplot as plt 
from math import pi
```

```python
10*(pi/180)
```

```python
start = 2*pi
end = (4*pi)+(pi/180)
step = 10*(pi/180)
t = np.arange(start,end,step)   
a = np.sin(t)+1             # Sin + 1 since we only want positive values
a = [i*2000 for i in a]     # Multiply by 2000 so its in the range we want
a = [int(i) for i in a]     # Convert to int
print(len(a))
fig,ax = plt.subplots(figsize=(10,7))
ax.plot(t,a)
plt.show()
```
