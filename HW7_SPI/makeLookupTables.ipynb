```python
import numpy as np 
from matplotlib import pyplot as plt 
from math import pi
```

# Sine wave

```python
start = 2*pi
end = (4*pi)+(pi/180)
step = 10*(pi/180)
t = np.arange(start,end,step)   
a = np.sin(t)+1             # Sin + 1 since we only want positive values
a = [i*2000 for i in a]     # Multiply by 2000 so its in the range we want
a = [int(i) for i in a]     # Convert to int
print(len(a))
fig,ax = plt.subplots()
ax.plot(t,a)
plt.show()
```

# Triangle wave
$ y = \frac{2a}{\pi} \cdot arcsin(sin(\frac{2 \pi}{p}x)) $  
With $p=2\pi$  

$ y = \frac{2a}{\pi} \cdot arcsin(sin(x)) $

```python

start = 2*pi
end = (4*pi)+(pi/180)
step = 10*(pi/180)
t = np.arange(start,end,step) 

y = (2*2000/pi)*np.arcsin(np.sin(t))+2000  # Sin + 2000 since we only want positive values in the range
y = [int(i) for i in y]                    # Convert to int
print(y)
fig,ax = plt.subplots()
ax.plot(t,y)
plt.show()

```
